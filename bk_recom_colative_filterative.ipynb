# Popularity Based Book Recommendation system
- base of hight avg rating 

1. merge the rating with book by ISBN
2. group by base on book-rating and reset index
3. create two table
 -  num_rating_df
 - avg_rating_df
4. merge num_rating_df and avg_rating_df as popular_df
5. find the top 50 book whose have avg_rating > 250 and sorted on asc order
6. find the books detail from books dataset
  - merge the popular_df dataset to books
  - remove the duplicated books who have same book-title

```python
# merge the ratings with books
rating_with_name=ratings.merge(books, on="ISBN")
```

```python
print(rating_with_name.shape)
print(rating_with_name.columns)
```

```python
rating_with_name.head(2)
```

```python
num_rating_df = rating_with_name.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_rating_df.rename(columns={"Book-Rating":"number-rating"}, inplace=True)
num_rating_df
```

```python
avg_rating_df = rating_with_name.groupby("Book-Title").mean()["Book-Rating"].reset_index()
avg_rating_df.rename(columns={"Book-Rating":"avg-rating"},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on="Book-Title")
popular_df.head(2)
```

```python
popular_df = popular_df[popular_df["number-rating"] > 250].sort_values("avg-rating",ascending=False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(books, on="Book-Title").drop_duplicates("Book-Title")
print(popular_df.shape)
popular_df.columns
```

```python
popular_df=popular_df[['Book-Title', 'number-rating', 'avg-rating', 'Book-Author','Image-URL-M',
       'Image-URL-S']]
```

# Collaborative Filtering Based Recommender System

1. 

```python
x = rating_with_name.groupby('User-ID').count()["Book-Rating"]>300
padh_likhe_user = x[x].index
padh_likhe_user
```

```python

filter_rating = rating_with_name[rating_with_name["User-ID"].isin(padh_likhe_user)]
filter_rating
```

```python
filter_rating.columns
```

```python
y = filter_rating.groupby("Book-Title").count()["Book-Rating"] >= 50
famous_book = y[y].index
```

```python
final_book = filter_rating[filter_rating["Book-Title"].isin(famous_book)] ## this is the way youtube go
final_book
```

```python
# Now create Pivot Table
pt = final_book.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating")
```

```python
pt.fillna(0, inplace=True)
```

```python
import pickle
pickle.dump(pt,open('pt_df.pkl','wb'))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key= lambda x:x[1], reverse=True)[1:6]
    
    data = []
    for i in similar_items:
      # print(pt.index[i[0]])
      item = []
      temp_df = books[books['Book-Title'] == pt.index[i[0]]]
      item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
      item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
      item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

      data.append(item)
    
    return data
```

```python
recommend("A Walk to Remember")
```

```python
recommend_book = recommend("A Walk to Remember")
```
